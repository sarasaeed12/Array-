```python
import qrcode
import cv2
from PIL import Image
from IPython.display import display

def generate_qr(data, filename="qrcode.png"):
    qr = qrcode.QRCode(version=1, box_size=10, border=5)
    qr.add_data(data)
    qr.make(fit=True)
    img = qr.make_image(fill_color="blue", back_color="white")
    img.save(filename)
    display(img)
    print(f"‚úÖ QR code saved as {filename}")

def decode_qr(filename):
    img = cv2.imread(filename)
    detector = cv2.QRCodeDetector()
    data, bbox, _ = detector.detectAndDecode(img)

    if data:
        print(f"üëç Decoded Data: {data}")
    else:
        print("üê±‚Äçüëì No QR code found, try using a white background.")
        generate_qr("Successfully generated", "my_qr.png")
        decode_qr("my_qr.png")

# Example usage
generate_qr("Hello world!", "test_qr.png")
decode_qr("test_qr.png")
```
